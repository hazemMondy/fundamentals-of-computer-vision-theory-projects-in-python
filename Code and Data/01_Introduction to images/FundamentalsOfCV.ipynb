# RGB to HSV conversion
color space: hue saturation contrast(luminance)


```python
plt.imshow(plt.imread(r'Screenshot.jpg'))
```

```python
fig , ax = plt.subplots()
plt.imshow(plt.imread(r'Screenshot 1.jpg'));
ax.axis('off');
```

```python
def f_rgb_to_hsv(r, g, b,scaleFactor): 
    # normalize rgb values
    r, g, b = r / 255.0, g / 255.0, b / 255.0
    # find the maximum and minimum rgb values
    cmax = max(r, g, b)    # maximum of r, g, b 
    cmin = min(r, g, b)    # minimum of r, g, b 
    diff = cmax-cmin       # diff of cmax and cmin. 
    

    if cmax == cmin:  
        h = 0
    elif cmax == r:  
        h = (60 * ((g - b) / diff) + 0) % 360
    elif cmax == g: 
        h = (60 * ((b - r) / diff) + 120) % 360
    elif cmax == b: 
        h = (60 * ((r - g) / diff) + 240) % 360
   
    if h < 0:
        h = h + 360
    if cmax == 0: 
        s = 0
    else: 
        s = (diff / cmax) * scaleFactor
    v = cmax * scaleFactor
    return h, s, v 
```

```python
print(f_rgb_to_hsv(100,200,50,100))
```

```python
img = cv2.imread(r'tulips.jpg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
x=1
# ! masking
# segmengation for red color
# in hsv in opnecv the red range (0:30), (150:180)
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
lr = np.array([0,120,70])
ur = np.array([10,255,255])
mask1 = cv2.inRange(hsv,lr,ur)
```

```python
type(mask1)
```

```python
mask1.dtype
```

```python
mask1.shape
```

```python
mask1.max()
```

```python
mask1.min()
```

```python
lr = np.array([170,120,70])
ur = np.array([180,255,255])
mask2 = cv2.inRange(hsv,lr,ur)
```

```python
np.unique(mask2)
```

```python
np.unique(mask1)
```

```python
mask = mask1 | mask2
```

```python
np.unique(mask)
```

```python
res = cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```

```python
# solution to quiz
lr = np.array([20,100,100])
ur = np.array([30,255,255])
yellow_mask = cv2.inRange(hsv,lr,ur)
res = cv2.bitwise_and(img,img,mask=yellow_mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
